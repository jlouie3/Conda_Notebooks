```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Using car evaluation data from UCI ML data repository
# http://archive.ics.uci.edu/ml/datasets/Car+Evaluation

# Tutorials:
#   https://seaborn.pydata.org/tutorial/distributions.html
#   https://www.datacamp.com/community/tutorials/seaborn-python-tutorial#load
#   https://elitedatascience.com/python-seaborn-tutorial

df = pd.read_csv('data/car.data')
df.head()  # show data
```

```python
def price_map(x):
    if x=='vhigh':
        result = 4.0/4
    elif x=='high':
        result = 3.0/4
    elif x=='med':
        result = 2.0/4
    elif x=='low':
        result = 1.0/4
    else:
        result = 0
        
    return result

def acceptability_map(x):
    if x=='vgood':
        result = 4.0/4
    elif x=='good':
        result = 3.0/4
    elif x=='acc':
        result = 2.0/4
    elif x=='unacc':
        result = 1.0/4
    else:
        result = 0
        
    return result

df['buyPrice'] = df['buyPrice'].apply(price_map)
df['maintPrice'] = df['maintPrice'].apply(price_map)
df['acceptability'] = df['acceptability'].apply(acceptability_map)
df.head()
```

```python
sns.jointplot(x='buyPrice', y='acceptability', data=df)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
